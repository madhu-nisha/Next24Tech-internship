```python
import pandas as pd
from twilio.rest import Client


def detect_heatwaves(data, threshold=30, consecutive_days=3):
    data['heatwave'] = False
    count = 0

    for i in range(len(data)):
        if data['Data.Temperature.Max Temp'][i] > threshold:
            count += 1
        else:
            count = 0

        if count >= consecutive_days:
            data['heatwave'][i] = True

    return data


def send_sms(to, message):
    account_sid = 'AC7aaabcab50ceb8643a1ac78be055deec'
    auth_token = '90ed63e8a26c0957ebebcf3a378cbf0b'
    client = Client(account_sid, auth_token)

    message = client.messages.create(
        body=message,
        from_='+916369770335',
        to=to
    )

    print(f"Message sent to {to}: {message.sid}")


# Load historical temperature data from a CSV file
data = pd.read_csv('/content/historical_temperature_data.csv')

# Apply heatwave detection
data = detect_heatwaves(data)

# Check for heatwaves and send notifications
for index, row in data.iterrows():
    if row['heatwave']:
        send_sms('+914864654682', f"Heatwave detected on {row['Date.Full']} with temperature {row['Data.Temperature.Max Temp']}°C.")
```
